```python
import pandas as pd

url = "https://raw.githubusercontent.com/nkmwicz/data-for-students/refs/heads/main/french-ambs.csv"
df = pd.read_csv(url)
df.head()
```

```python
df.shape
```

```python
filtered = df[(df["year"] > 1588) & (df["year"] < 1592)]
filtered.shape
```

```python
# Create column with default true
df["is_permenant"] = True

# Set false for temp embassies
temporary_places = ["Geneva", "Saxony", "Tuscany", "Ferrara"]

df.loc[df["place"].isin(temporary_places), "is_permenant"] = False

# Count the false values
df.loc[df["is_permenant"] == False].shape
```

```python
df["place"] = df["place"].replace(["Geneva", "Grisons"], "Swiss Cantons")
```

```python
(df["place"] == "Swiss Cantons").sum()
```

```python
df_agg = df.groupby("place").agg(
    years_present=("place", "count"),
    unique_ambs=("name", "nunique")
).reset_index()

# Sort ascending by unique_ambs
df_agg = df_agg.sort_values(by="unique_ambs", ascending=True)

df_agg
```
